# Robust Local Plane Gradient step-edge example

A nonlinear local plane fit with robust window weighting. This notebook creates a simple black-to-white step edge, runs the filter, and plots the result.


## How this filter works

This filter fits a small local plane to the image and uses the plane slopes as `gradient_x` and `gradient_y`. Robust weights such as Huber, Tukey, or bilateral-style weights reduce the influence of outliers and pixels that do not look like the center pixel.

### Paper reference

This exact local-plane combination is package-specific. The weighting pieces are tied to Huber's 1964 robust-estimation paper, DOI [10.1214/aoms/1177703732](https://doi.org/10.1214/aoms/1177703732), Beaton and Tukey's 1974 biweight fitting paper, DOI [10.1080/00401706.1974.10489171](https://doi.org/10.1080/00401706.1974.10489171), Tomasi and Manduchi's 1998 bilateral-filtering paper, DOI [10.1109/ICCV.1998.710815](https://doi.org/10.1109/ICCV.1998.710815), and Cleveland's 1979 local-regression paper, DOI [10.1080/01621459.1979.10481038](https://doi.org/10.1080/01621459.1979.10481038). Links. [Robust regression](https://en.wikipedia.org/wiki/Robust_regression), [Huber loss](https://en.wikipedia.org/wiki/Huber_loss), and [bilateral filter](https://en.wikipedia.org/wiki/Bilateral_filter).


## Setup

Import the small set of tools used in this notebook. The autoreload lines help Jupyter pick up local source edits without restarting the kernel every time.


```python
# Imports and local reloads.
%load_ext autoreload
%autoreload 2

import time

import matplotlib.pyplot as plt
import torch

from agfb_filters import ExecutionPath, get_filter_definition, run_filter
```

## Filter settings

Choose the filter settings here. The `definition` describes the filter, and the `path` chooses the matching execution path.


```python
# Change these settings to try the filter.
radius = 1
weighting = "huber"
definition = get_filter_definition(
    "robust_local_plane_gradient", radius=radius, weighting=weighting
)
path = ExecutionPath.NONLINEAR_WINDOW
```

## Test image

Create a test image. By default this is a synthetic step edge. The commented block lets you replace it with your own image file path.


```python
# Build the step-edge test image.
size = 1024
image = torch.zeros(1, size, size)
image[:, :, size // 2 :] = 1.0

# To use your own image instead, uncomment this block and set image_path.
# image_path = "path/to/image.png"
# image = torch.as_tensor(plt.imread(image_path), dtype=torch.float32)
# if image.ndim == 3:
#     image = image[..., :3].mean(dim=-1)
# image = image.unsqueeze(0)

plt.figure(figsize=(4, 4))
plt.imshow(image[0], cmap="gray", vmin=0, vmax=1)
plt.title("input image")
plt.axis("off")
```

## Run the filter

Run the filter. The timer measures only the filter call, not the image setup or plotting.


```python
# Time the filter call.
start = time.perf_counter()
gradient_x, gradient_y = run_filter(
    definition,
    image,
    path=path,
    boundary=definition.default_boundary,
)
elapsed = time.perf_counter() - start

print(f"{elapsed:.4f} seconds")
```

## Gradient results

Plot the two gradient outputs with the same grayscale range. The input changes left-to-right, so the strongest response should usually be in `gradient_x`.


```python
# Plot gradient_x and gradient_y.
limit = max(float(gradient_x.abs().max()), float(gradient_y.abs().max()))

fig, axes = plt.subplots(1, 2, figsize=(8, 3))

axes[0].imshow(gradient_x[0], cmap="gray", vmin=-limit, vmax=limit)
axes[0].set_title("gradient_x")

axes[1].imshow(gradient_y[0], cmap="gray", vmin=-limit, vmax=limit)
axes[1].set_title("gradient_y")

for ax in axes:
    ax.axis("off")

fig.suptitle(f"{definition.name}, time={elapsed:.4f} s")
fig.tight_layout()
```
